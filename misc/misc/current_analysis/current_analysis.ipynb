```python
import pandas as pd
import matplotlib.pyplot as plt

# %matplotlib qt
```

```python
col_name = ['values']
data_3N = pd.read_csv('putty_3N.csv', names=col_name)
data_4N = pd.read_csv('putty_4N.csv', names=col_name)
data_5N = pd.read_csv('putty_5N.csv', names=col_name)
data_6N = pd.read_csv('putty_6N.csv', names=col_name)
data_7N = pd.read_csv('putty_7N.csv', names=col_name)
data_8N = pd.read_csv('putty_8N.csv', names=col_name)
data_10N = pd.read_csv('putty_10N.csv', names=col_name)
data_15N = pd.read_csv('putty_15N.csv', names=col_name)
data_20N = pd.read_csv('putty_20N.csv', names=col_name)

data_3N.head()
data_4N.head()
```

```python

data_3N_np = (data_3N.to_numpy()).flatten()
data_4N_np = (data_4N.to_numpy()).flatten()
data_5N_np = (data_5N.to_numpy()).flatten()
data_6N_np = (data_6N.to_numpy()).flatten()
data_7N_np = (data_7N.to_numpy()).flatten()
data_8N_np = (data_8N.to_numpy()).flatten()
data_10N_np = (data_10N.to_numpy()).flatten()
data_15N_np = (data_15N.to_numpy()).flatten()
data_20N_np = (data_20N.to_numpy()).flatten()


data_list = [data_3N_np, data_4N_np, data_5N_np, data_6N_np, data_7N_np, data_8N_np, data_10N_np, data_15N_np, data_20N_np]
labels = ['3N', '4N', '5N', '6N', '7N', '8N', '10N', '15N', '20N']
time = 1000/5

time_data_list = []

for data in data_list:
    time_data = []
    for i in range(len(data)):
        time_data.append(i/time)
    time_data_list.append(time_data)

print(len(data_3N_np))
print(len(data_4N_np))
```

```python

nb_rows = 9
fig, ax = plt.subplots(nb_rows, 1, figsize=(10, 20))

for i in range(nb_rows):
    ax[i].set_xlabel('Time [s]')

for i in range(nb_rows):
    ax[i].plot(time_data_list[i], data_list[i], label=labels[i])

for i in range(nb_rows):
    ax[i].set_ylim(0, 50)
    ax[i].set_ylabel('Current [A]')
    ax[i].legend()

# for i in range(nb_rows - 1):
#     # remove xticks
#     plt.setp(ax[i].get_xticklabels(), visible=False)
plt.savefig('measurements.png')
plt.show()
```
