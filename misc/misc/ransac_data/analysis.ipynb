```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

header = ['base_link_x', 'base_link_y', 'base_link_z', \
          'base_link_vx', 'base_link_vy', 'base_link_vz', \
          'right_foot_x', 'right_foot_y', 'right_foot_z', \
          'right_foot_vx', 'right_foot_vy', 'right_foot_vz', \
          'left_foot_x', 'left_foot_y', 'left_foot_z',
          'left_foot_vx', 'left_foot_vy', 'left_foot_vz']

df = pd.read_csv('refined_translation.csv', names=header)
df.head()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].plot(df['base_link_x'][10:],'*', label='base_link_x')
ax[0].plot(df['base_link_y'][10:],'*', label='base_link_y')
ax[0].plot(df['base_link_z'][10:],'*', label='base_link_z')
ax[0].set_title('Base Link Delta Position')
ax[0].legend()
ax[1].plot(df['base_link_vx'][10:],'*', label='base_link_vx')
ax[1].plot(df['base_link_vy'][10:],'*', label='base_link_vy')
ax[1].plot(df['base_link_vz'][10:],'*', label='base_link_vz')
ax[1].set_title('Base Link Delta Orientation')
plt.legend()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].plot(df['right_foot_x'][10:],'*', label='right_foot_x')
ax[0].plot(df['right_foot_y'][10:],'*', label='right_foot_y')
ax[0].plot(df['right_foot_z'][10:],'*', label='right_foot_z')
ax[0].set_title('Right Foot Delta Position')
ax[0].legend()
ax[1].plot(df['right_foot_vx'][10:],'*', label='right_foot_vx')
ax[1].plot(df['right_foot_vy'][10:],'*', label='right_foot_vy')
ax[1].plot(df['right_foot_vz'][10:],'*', label='right_foot_vz')
ax[1].set_title('Right Foot Delta Orientation')
plt.legend()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].plot(df['left_foot_x'][10:],'*', label='left_foot_x')
ax[0].plot(df['left_foot_y'][10:],'*', label='left_foot_y')
ax[0].plot(df['left_foot_z'][10:],'*', label='left_foot_z')
ax[0].set_title('Left Foot Delta Position')
ax[0].legend()
ax[1].plot(df['left_foot_vx'][10:],'*', label='left_foot_vx')
ax[1].plot(df['left_foot_vy'][10:],'*', label='left_foot_vy')
ax[1].plot(df['left_foot_vz'][10:],'*', label='left_foot_vz')
ax[1].set_title('Left Foot Delta Orientation')
plt.legend()
```
