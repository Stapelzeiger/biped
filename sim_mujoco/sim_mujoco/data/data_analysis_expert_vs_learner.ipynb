```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#FREQUENCY 1000 Hz

dataset_no_spectral = pd.read_csv('dataset_comparison_no_spectral.csv')
dataset_no_spectral.head()

dataset_spectral = pd.read_csv('dataset_comparison_spectral.csv')
dataset_spectral.head()

adjusted_time = dataset_no_spectral['time'] - dataset_no_spectral['time'][0]
adjusted_time_spectral = dataset_spectral['time'] - dataset_spectral['time'][0]

```

```python
# plot all the joints
list_joints_names = ['YAW', 'HAA', 'HFE', 'KFE', 'ANKLE']



fig, axs = plt.subplots(2, 2, figsize=(10, 10))
for joint_name in list_joints_names:
    axs[0,0].set_title('left joints pos')
    axs[0,0].plot(adjusted_time, dataset_no_spectral['L_'+ joint_name + '_pos'])
    axs[0,1].set_title('right joints pos')
    axs[0,1].plot(adjusted_time, dataset_no_spectral['R_'+ joint_name + '_pos'])

    axs[1,0].set_title('left joints vel')
    axs[1,0].plot(adjusted_time, dataset_no_spectral['L_'+ joint_name + '_vel'])
    axs[1,1].set_title('right joints vel')
    axs[1,1].plot(adjusted_time, dataset_no_spectral['R_'+ joint_name + '_vel'])

# print(len(dataset['L_'+ joint_name + '_pos']))
```

```python

plt.plot(dataset_no_spectral['vel_x_BF'], label='x')
plt.plot(dataset_no_spectral['vel_y_BF'], label='y')
plt.plot(dataset_no_spectral['vx_des_BF'], label='x_des')
plt.plot(dataset_no_spectral['vy_des_BF'], label='y_des')
plt.legend()
# Note, see the fact that we walk with higher velocity at some point (aka not walk backwards lol)
```

```python

plt.plot(dataset_no_spectral['normal_vec_x_BF'], label='x')
plt.plot(dataset_no_spectral['normal_vec_y_BF'], label='y')
plt.plot(dataset_no_spectral['normal_vec_z_BF'], label='z')
plt.legend()
```

```python
# plot all the joints
list_joints_names = ['YAW', 'HAA', 'HFE', 'KFE']

fig, axs = plt.subplots(len(list_joints_names), 2, figsize=(10, 10))
for i, joint_name in enumerate(list_joints_names):
    axs[i,0].set_title('L_'+ joint_name + '_q_des')
    axs[i,0].plot(adjusted_time, dataset_no_spectral['L_'+ joint_name + '_q_des'], label='expert')
    axs[i,0].plot(adjusted_time, dataset_no_spectral['L_' + joint_name + '_q_des_policy'], label='NN_no_spectral')
    axs[i,0].plot(adjusted_time_spectral, dataset_spectral['L_' + joint_name + '_q_des_policy'], label='NN_with_spectral')
    axs[i,0].legend()
    axs[i,1].set_title('R_'+ joint_name + '_q_des')
    axs[i,1].plot(adjusted_time, dataset_no_spectral['R_'+ joint_name + '_q_des'],  label='expert')
    axs[i,1].plot(adjusted_time, dataset_no_spectral['R_' + joint_name + '_q_des_policy'], label='NN_no_spectral')
    axs[i,1].plot(adjusted_time_spectral, dataset_spectral['R_' + joint_name + '_q_des_policy'], label='NN_with_spectral')
    axs[i,1].legend()
plt.tight_layout()


fig, axs = plt.subplots(len(list_joints_names), 2, figsize=(10, 10))
for i, joint_name in enumerate(list_joints_names):
    axs[i,0].set_title('L_'+ joint_name + '_q_vel_des')
    axs[i,0].plot(adjusted_time, dataset_no_spectral['L_'+ joint_name + '_q_vel_des'], label='expert')
    axs[i,0].plot(adjusted_time, dataset_no_spectral['L_' + joint_name + '_q_vel_des_policy'], label='NN_no_spectral')
    axs[i,0].plot(adjusted_time_spectral, dataset_spectral['L_' + joint_name + '_q_vel_des_policy'], label='NN_with_spectral')
    axs[i,0].legend()

    axs[i,1].set_title('R_'+ joint_name + '_q_vel_des')
    axs[i,1].plot(adjusted_time, dataset_no_spectral['R_'+ joint_name + '_q_vel_des'],  label='expert')
    axs[i,1].plot(adjusted_time, dataset_no_spectral['R_' + joint_name + '_q_vel_des_policy'], label='NN_no_spectral')
    axs[i,1].plot(adjusted_time_spectral, dataset_spectral['R_' + joint_name + '_q_vel_des_policy'], label='NN_with_spectral')
    axs[i,1].legend()
plt.tight_layout()


fig, axs = plt.subplots(len(list_joints_names), 2, figsize=(10, 10))
for i, joint_name in enumerate(list_joints_names):
    axs[i,0].set_title('L_'+ joint_name + '_tau_ff')
    axs[i,0].plot(adjusted_time, dataset_no_spectral['L_'+ joint_name + '_tau_ff'], label='expert')
    axs[i,0].plot(adjusted_time, dataset_no_spectral['L_' + joint_name + '_tau_ff_policy'], label='NN_no_spectral')
    axs[i,0].plot(adjusted_time_spectral, dataset_spectral['L_' + joint_name + '_tau_ff_policy'], label='NN_with_spectral')

    axs[i,0].legend()

    axs[i,1].set_title('R_'+ joint_name + '_tau_ff')
    axs[i,1].plot(adjusted_time, dataset_no_spectral['R_'+ joint_name + '_tau_ff'],  label='expert')
    axs[i,1].plot(adjusted_time, dataset_no_spectral['R_' + joint_name + '_tau_ff_policy'], label='NN_no_spectral')
    axs[i,1].plot(adjusted_time_spectral, dataset_spectral['L_' + joint_name + '_tau_ff_policy'], label='NN_with_spectral')
    axs[i,1].legend()

plt.tight_layout()
```
